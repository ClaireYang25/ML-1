```
# Boston Housing price: used for assessment of houses prices in Boston.
#  https://www.ccs.neu.edu/home/vip/teach/MLcourse/data/housing_desc.txt 
#  x (input features) contains (506, 13) values divided into training and testing sets, 
#  y (target feature) is Median value of owner-occupied homes in $1000's

import numpy as np
import matplotlib.pyplot as plt
from keras.datasets.boston_housing import load_data
```

```
(xtrain, ytrain), (xtest, ytest) = load_data()
print(xtrain.shape)
print(xtest.shape)

t0 = np.arange(0, ytrain.shape[0])
t1 = np.arange(ytrain.shape[0], ytrain.shape[0]+ytest.shape[0])


plt.figure
plt.plot(t0, ytrain,'b-')
plt.plot(t1, ytest,'r-')
plt.grid()
plt.show()

```
