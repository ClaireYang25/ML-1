```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.arange(-6.0,2.0,0.1)

def f(x):
  f = x**4 + 7 * x**3 + 5 * x**2 - 17 * x + 3
  return(f)

def df(x):
  df = 4 * x**3 + 21 * x**2 + 10 * x - 17
  return(df)

def d2f(x):
  d2f = 12 * x**2 + 42 * x + 10
  return(d2f)


x_star =np.array([-4.5, 0.7, -1.4])

print(x_star)
print(f(x_star))
print(d2f(x_star))

fig = plt.figure()
plt.plot(x,f(x),'b-')
plt.plot(x,df(x),'g-')
plt.plot(x,d2f(x),'r-')
plt.plot(x_star, f(x_star), 'ro', markersize=10)
plt.grid()
plt.xlabel('x')
plt.ylabel('f')
plt.show()
```

```
# implement gradient descent for univariate function

def f(x):
  f = x**4 + 7 * x**3 + 5 * x**2 - 17 * x + 3
  return(f)

def df(x):
  df = 4 * x**3 + 21 * x**2 + 10 * x - 17
  return(df)


eta = 1e-3
alpha = 1e-2
max_iter = 100
x = np.zeros(max_iter)

iteration = 0
x[iteration] = 2.0

while iteration < max_iter-1:
  x[iteration+1] = x[iteration] - eta * df(x[iteration]) + alpha * (x[iteration]-x[iteration-1])
  iteration += 1

plt.figure()
plt.plot(np.arange(-6,2, 0.01), f(np.arange(-6,2, 0.01)), 'b-')
plt.plot(x, f(x), 'r*')
plt.grid()
plt.show()
```
